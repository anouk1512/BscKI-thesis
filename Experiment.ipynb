```python
def plot_two_radars(vector_1, vector_2, title_1, title_2, total_title): 
    fig = go.Figure()
    categories = old_competences
    
    fig.add_trace(go.Scatterpolar(
          r=vector_1,
          theta=categories,
          fill='toself',
          name=title_1
    ))
    fig.add_trace(go.Scatterpolar(
          r=vector_2,
          theta=categories,
          fill='toself',
          name=title_2
    ))

    fig.update_layout(title_text=total_title, title_x=0.5, legend_traceorder='reversed',
    title_font = dict(size =20, color='black', family='Arial, sans-serif'), 
      polar=dict(
        radialaxis=dict(
          visible=True,
          range=[0, 1]
        )),
      showlegend=True
    )

    fig.show()
    return 


def plot_two_vectors(vector_1, vector_2, title_1, title_2, total_title): 
    categories = old_competences
    categories = [*categories, categories[0]]

    first = vector_1
    second = vector_2
    first = [*first, first[0]]
    second = [*second, second[0]]

    fig = go.Figure(
        data=[
            go.Scatterpolar(r=first, theta=categories, fill='toself', name=title_1),
            go.Scatterpolar(r=second, theta=categories, fill='toself', name=title_2),
        ],
        layout=go.Layout(
            title=go.layout.Title(text=total_title),
            polar=dict(
        radialaxis=dict(
          visible=True,
          range=[0, 1]
        )),
            showlegend=True
        )
    )
    fig.update_layout(legend_font_size=20, title_font_size=20, title_x=0.5, legend_traceorder='reversed', 
                     )
    fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="right",
    x=0.85
))
    pyo.plot(fig)
    return
```
